```python
from pathlib import Path
import pandas as pd

# Opciones de display para ver dataframes MUY anchos
pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 180)
pd.set_option("display.max_colwidth", 120)

# Directorios
RAW_DIR = Path("../data/raw").resolve()

RAW_DIR, list(RAW_DIR.glob("*"))
```

```python
# Elegí por patrón parte del nombre (insensible a mayúsculas)
patron = "censal"  # ejemplo: "cc", "lengua", "secundaria", etc.

candidatos = [p for p in RAW_DIR.glob("*") if patron.lower() in p.name.lower()]
candidatos
```

```python
# Tomo el primero de la lista (o el que elijas por índice)
archivo = candidatos[0]
archivo, archivo.suffix
```

```python
def leer_archivo(path: Path):
    if path.suffix.lower() in [".csv", ".txt"]:
        # Intento 1: decimales con punto
        try:
            df = pd.read_csv(
                path,
                sep=",",
                thousands=",",
                decimal=".",
                low_memory=False,
                encoding="utf-8-sig",
            )
            return df
        except Exception as e1:
            # Intento 2: decimales con coma
            try:
                df = pd.read_csv(
                    path,
                    sep=",",
                    thousands=".",
                    decimal=",",
                    low_memory=False,
                    encoding="latin-1",
                )
                return df
            except Exception as e2:
                raise RuntimeError(f"No pude leer el CSV.\nIntento1: {e1}\nIntento2: {e2}")
    elif path.suffix.lower() in [".xlsx", ".xls"]:
        return pd.read_excel(path, engine="openpyxl")
    else:
        raise ValueError(f"Extensión no soportada: {path.suffix}")

df = leer_archivo(archivo)
df.shape, df.columns[:10].tolist()
```

```python
# Tamaño y tipos
print("Shape:", df.shape)
print("\nTipos (primeros 20):")
print(df.dtypes.head(20))

# Primeras filas
display(df.head(3))

# Nulos por columna (top 20)
nul = df.isna().sum().sort_values(ascending=False).head(20)
nul.to_frame("n_missing")
```

```python
def buscar_columnas(patrones):
    patrones = [p.lower() for p in patrones]
    cols = [c for c in df.columns if any(p in c.lower() for p in patrones)]
    return cols

# Ajustá estos patrones según lo que encuentren en el diccionario de variables
cols_apoyo = buscar_columnas(["apoyo", "famil", "hogar", "responsa", "nivel_ed_persona_resp"])
cols_aspir = buscar_columnas(["aspira", "estudiar", "trabajar", "universidad", "secundaria"])
cols_rend  = buscar_columnas(["lengua", "matem", "puntaje", "desempe", "tri", "nivel"])

len(cols_apoyo), len(cols_aspir), len(cols_rend), cols_rend[:15]
```
