```python
from collections import Counter

# tu mapping (ejemplo: dfrenombre tiene 'Columnas 2019' -> 'Renombre_2019')
mapping = dict(zip(dfrenombre['Columnas 2019'], dfrenombre['Renombre_2019']))

# claves del mapping que existen en el dataframe
claves_presentes = set(mapping.keys()) & set(df1.columns)
claves_faltantes = set(mapping.keys()) - set(df1.columns)

print("Claves del mapping presentes en df1:", len(claves_presentes))
print("Claves del mapping NO encontradas en df1:", len(claves_faltantes))
print("Ejemplos faltantes:", list(claves_faltantes)[:10])

# detectar colisiones (varios 'old' que apuntan al mismo 'new')
nv = list(mapping.values())
dups_new = [name for name, c in Counter(nv).items() if c > 1]
print("Nuevos nombres duplicados (colisiones):", dups_new[:20])

# para cada nuevo nombre duplicado, mostrar las columnas antiguas que colisionan
from collections import defaultdict
colisiones = defaultdict(list)
for old, new in mapping.items():
    if new in dups_new:
        colisiones[new].append(old)

for new, olds in colisiones.items():
    print(f"{new}: {olds}")
```

```python
def guardar_df(df: pd.DataFrame, ruta: str):
    """
    Guarda un DataFrame en un archivo CSV en la ruta indicada.

    Parámetros:
    df (pd.DataFrame): DataFrame a guardar.
    ruta (str): Ruta completa del archivo donde se guardará el CSV.
    """
    try:
        df.to_csv(ruta, index=False, encoding='utf-8')
        print(f"Archivo guardado exitosamente en: {ruta}")
    except Exception as e:
        print(f"Error al guardar el archivo: {e}")
```
