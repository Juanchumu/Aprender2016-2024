# Apertura 2024

se buscan coincidencias y se unifican

```python
df3 = pd.read_csv(ruta_archivo_2024, sep=',', decimal=".", low_memory=False)
```

```python
# 1. Encontrar columnas comunes a los 3 df
cols_comunes = (
    (set(df1.columns)
    & set(df2.columns))
    & set(df3.columns)
)

print(len(set(df1.columns)))
print(len(set(df2.columns)))
print(len(set(df3.columns)))

print(len(cols_comunes))
#print(cols_comunes)
```

```python
cols_comunes = list(cols_comunes)   # convertir a lista para indexar

# 2. Filtrar cada df a solo columnas comunes
df1_c = df1[cols_comunes].copy()
df2_c = df2[cols_comunes].copy()
df3_c = df3[cols_comunes].copy()

# 3. Concatenar verticalmente
df_final = pd.concat([df1_c, df2_c, df3_c], axis=0, ignore_index=True)

```
